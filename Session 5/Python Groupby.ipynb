```python
import pandas as pd
```

```python
df = pd.read_csv('Data.csv')
df.head()
```

```python
df = df.drop(columns='Total')
```

```python
df.info()
```

```python
df['Total'] = df['Units'] * df['Unit Cost']
```

```python
df.head()
```

```python
df.groupby('Rep').sum()
```

```python
df.groupby('Rep', as_index=False).sum()
```

```python
df.groupby('Rep', as_index=False)['Total'].sum()
```

```python
df.groupby('Rep', as_index=False)['Total'].agg('mean')
```

```python
df.groupby('Rep')['Total'].agg(['sum', 'mean', 'max'])
```

```python
df.groupby('Rep')[['Total', 'Unit Cost']].agg({'Total': ['max', 'min'], 'Unit Cost': ['mean', 'sum']})
```

```python
df.groupby(['Rep', 'Region']).mean()
```
