```python
import matplotlib.pyplot as plt
```

```python
# x-axis values 
x = [5, 2, 9, 4, 7] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot 
plt.plot(x,y) 
  
# function to show the plot 
plt.show()
```

```python
# x-axis values 
x = [5, 2, 9, 4, 7] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot the bar 
plt.bar(x,y) 
  
# function to show the plot 
plt.show() 
```

```python
from matplotlib import pyplot as plt 
  
# Y-axis values 
y = [1, 2, 3, 4, 5, 1, 2, 1, 6, 7, 8, 9, 10] 
  
# Function to plot histogram 
plt.hist(y) 
  
# Function to show the plot 
plt.show()
```

```python
 y = [1,2,3,4,5,1,2,6,8,4,7,1,3, 3.5]
plt.hist(y)
plt.show()
```

```python
# x-axis values 
x = [5, 2, 9, 4, 7] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot scatter 
plt.scatter(x, y) 
  
# function to show the plot 
plt.show()
```
